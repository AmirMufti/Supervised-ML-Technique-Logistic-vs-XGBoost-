```python
# importing the visulation library
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Creating a list of columns to be called in the loop
columns = [
    'sessions', 'drives', 'total_sessions', 'n_days_after_onboarding',
    'total_navigations_fav1', 'total_navigations_fav2',
    'driven_km_drives', 'duration_minutes_drives',
    'activity_days', 'driving_days'
]
```

```python
# Defining functions for histogram and boxplots
def create_histogram(data, column_name):
    plt.figure(figsize=(4, 1))
    sns.histplot(x=data[column_name], color='#20E2D7', edgecolor='black')
    plt.title(f'Histogram of {column_name}')
    median = data[column_name].median()
    plt.axvline(median, color='red', linestyle='-')
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

def create_boxplot(data, column_name):
    plt.figure(figsize=(4, 1))
    sns.boxplot(x=data[column_name],fliersize=1, color='#20E2D7')
    plt.title(f'Boxplot of {column_name}')
    plt.ylabel(column_name)
    plt.grid(True)
    plt.show()
```

```python
# Defining function with for loop for generating all the plots
def generate_plots(data, columns):
    for n in columns:
        create_histogram(data, n)
        create_boxplot(data, n)
```

```python
generate_plots(df, columns)
```

```python
# Creating threshold and removing the outliers
def outliers(column_name, percentile):
    threshold = df[column_name].quantile(percentile)
    df.loc[df[column_name] > threshold, column_name] = threshold
    print('{:>25} | percentile: {} | threshold: {}'.format(column_name, percentile, threshold))
```

```python
for column in ['sessions', 'drives', 'total_sessions',
               'driven_km_drives', 'duration_minutes_drives']:
            outliers(column, 0.95)
```

```python
# Importing XGBoost and related packages/libraries
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.model_selection import GridSearchCV
```

```python
# Recreating the X and y dataframes
X = df.drop(columns = ['label', 'label2', 'device', 'sessions', 'driving_days'])
y = df['label2']
```

```python
# Splitting the new dataframe in to train and test datasets
X_train, X_test, y_train, y_test = train_test_split (X, y, random_state=42, stratify=y, test_size=0.2)
```

```python
# Instantiating the XGBClassifier 
xgb_model = XGBClassifier(objective = 'binary:logistic', random_state=42)
```

```python
# Defining parameters
parameters = {'max_depth': [6, 9],
             'min_child_weight': [4, 6],
             'learning_rate': [0.1, 0.15],
             'n_estimators': [400]
             }
```

```python
# Creating list of scores
score = ['accuracy', 'precision', 'recall', 'f1']
```

```python
xgb_tr = GridSearchCV(xgb_model, parameters, scoring = score, cv = 5, refit= 'recall')
```

```python
%%time
xgb_tr.fit(X_train, y_train)
```

```python
# Identifying the best parameters for the model
xgb_tr.best_params_
```

```python
# Predicting on the test dataset with the best parameters
xgb_preds = xgb_tr.best_estimator_.predict(X_test)
```

```python
# Calculate precision
precision = precision_score(y_test, xgb_preds)
print(f'Precision: {precision}')
```

```python
# Calculate recall 
recall = recall_score(y_test, xgb_preds)
print(f'Recall: {recall}')
```

```python
# Calculating accuracy
accuracy = accuracy_score(y_test, xgb_preds)
print(f'Accuracy: {accuracy}')
```

```python
# Calculating F1 Score
f1 = f1_score(y_test, xgb_preds)
print(f'F1 Score: {f1}')
```

```python
# Plotting the confusion matrix to identfy true positive and true negatives

cm = confusion_matrix(y_test, xgb_preds, labels=xgb_tr.classes_)

# Plot confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                             display_labels=['retained', 'churned'])
disp.plot();
```

```python
# Plotting the important features involved in predicting the churn rate
plot_importance(xgb_tr.best_estimator_);
```
